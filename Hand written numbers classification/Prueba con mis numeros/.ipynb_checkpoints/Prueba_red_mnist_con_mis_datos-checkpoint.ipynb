# Definiendo parametros estructurales de la red neuronal

Vamos a armar una red con 4 capas, 1 de entrada, dos intermedias y una de salida

```python
#Parametros de la estructura de nuestra red
numero_entradas = 784 # Numero de entrada que tendremos para nuestra primera capa 
neuronas_capa1 = 100 # Numero de neuronas en nuestra primera capa
neuronas_capa2 = 100 # Numero de neuronas en la segunda capa
numero_clasificaciones = 10 # Numero de clasificaciones en nuestra ultima capa
```

Defino las variables de TENSORFLOW que voy a usar

```python
#Matriz de entrada (784)
X = tf.placeholder("float", [None, numero_entradas])

#variable para pesos que van desde la entrada (784) hasta capa 1 (100)
pesos_entrada_capa1=tf.Variable(tf.random_normal([numero_entradas, neuronas_capa1])) #pesos de la entrada a la primera capa
#bias ("sesgos") para la primera capa
bias_capa1=tf.Variable(tf.random_normal([neuronas_capa1]))  #sesgo de primera capa

#variable para pesos que van desde la capa1 (100) hasta capa 2 (100)
pesos_capa1_capa2=tf.Variable(tf.random_normal([neuronas_capa1, neuronas_capa2])) #pesos de primera a segunda capa 
#bias segunda capa
bias_capa2=tf.Variable(tf.random_normal([neuronas_capa2])) #sesgo de segunda capa

#variable para pesos que van desde capa 2 (100) hasta la salida (10)
pesos_capa2_salida=tf.Variable(tf.random_normal([neuronas_capa2, numero_clasificaciones])) #pesos de segunda capa a salida
#bias de la ultima capa
bias_salida=tf.Variable(tf.random_normal([numero_clasificaciones])) #sesgo de ultima capa 

#Matriz de salida (10)
Y = tf.placeholder("float", [None, numero_clasificaciones])
```

Funcion para armar Red con los parametros antes definidos; el parametro que se le ingresa es el placeholder con los pixeles de entrada (784)

```python
#Funcion la cual une todo lo que hicimos en los ultimos pasos para ya tener una estructura armada 
def armar_red_neuronal(x):
    capa1 = tf.add(tf.matmul(x, pesos_entrada_capa1), bias_capa1)
    capa2 = tf.add(tf.matmul(capa1, pesos_capa1_capa2), bias_capa2)
    capa_Salida = tf.matmul(capa2, pesos_capa2_salida) + bias_salida
    return capa_Salida
```

```python
#Construimos la red y hacemos que la ultima capa pase por una funcion de softmax para hacer la clasificacion#Constru 
RN = armar_red_neuronal(X) #Le alimentamos X ya que es la variable que utilizaremos para la primera capa
prediccion = tf.nn.softmax(RN) #Cual es la prediccion de nuestra red (Que clasificacion)
```

# Entrenamiento de la red

Parametros de entrenamiento de nuestra red

```python
numero_pasos = 2000 #El numero de pasos (iteraciones) que se llevaran a cabo en el entrenamiento
lotes = 256 #Cada lote de entrenamiento de cuantos elementos va a ser
taza_aprendizaje = 0.01 #Que tan rapido aprende el algoritmo 
display_step = 100 #Cada cuantos pasos de entrenamiento nos desplegara la terminal como va el aprendizaje del programa
```

Creando las funciones las cuales se encargan de hacer que nuestro algoritmo aprenda

```python
perdida = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=RN, labels=Y))
optimizador = tf.train.AdamOptimizer(learning_rate=taza_aprendizaje)
minimizador = optimizador.minimize(perdida)
```

```python
#Funciones para medir que tan bien esta aprendiendo el programa durante su entrenamiento
predicciones = tf.equal(tf.argmax(prediccion, 1), tf.argmax(Y, 1))
certeza = tf.reduce_mean(tf.cast(predicciones, tf.float32))
```

Inicializacion de variables y session

```python
#inicializamos nuestras variables
init = tf.global_variables_initializer()
#session
sess=tf.Session()
sess.run(init)
```

Entrenamiento de nuestro modelo; lo hago de a pasos (500) y separando en lotes de 128 elementos cada uno

```python

for step in range(1, numero_pasos + 1):
    batch_x, batch_y = mnist.train.next_batch(lotes) #estos son los lotes que se van creando aleatoriamente

    sess.run(minimizador, feed_dict={X: batch_x, Y: batch_y}) 

    #Todo esto nos ense√±a que tan bien esta aprendiendo el programa cada 100 pasos de entrenamiento
    if step % display_step == 0 or step == 1:
        loss, acc = sess.run([perdida, certeza], feed_dict={X: batch_x,Y: batch_y})
        print("Paso "+str(step)+ ",Perdida= "+"{:.4f}".format(loss)+", Certeza= "+"{:.3f}".format(acc))
```

```python
print("Porcentaje de certeza Total: {:.2f}" .format(sess.run(certeza, feed_dict=
                                                       {X: mnist.test.images,Y: mnist.test.labels})*100),'%')
```
