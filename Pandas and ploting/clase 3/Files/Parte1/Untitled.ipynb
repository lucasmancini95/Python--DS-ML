```python
#CREO UN DF para practicar merge
names=['Martin','Jose','Rocio']
df_m1=pd.DataFrame({'Name':names,'sex':['M','M','F'], 'job':['L','S','T'], 'sport':['Rugby','Tennis','Rugby']})

df_m1
```

```python
names=['Martin','Sol','Luz']
df_m2=pd.DataFrame({'Name':names,'sex':['M','F','F'], 'job':['L','S','T'], 'sport':['Rugby','Tennis','Polo'], 'color':['red','blue','green']})

df_m2
```

```python
pd.merge(df_m1,df_m2,how='left') #me agarro los del df_m1 y le agrego la columna color del df_2, 
#puso los datos en los casos que correspondia
```

```python
pd.merge(df_m1,df_m2,how='right')
```

```python
pd.merge(df_m1,df_m2,how='outer')
```

```python
pd.merge(df_m1,df_m2,how='inner',left_on='sport',right_on='Sport')
```

```python
pd.merge(df_m1,df_m2,on=['job','sport'])
```

```python
df_m3=pd.DataFrame({'Otro':['a','b','c']})
frames=[df_m1,df_m2,df_m3] #lista de data frames
a=pd.concat(frames) # concatenacion de data frames
```
