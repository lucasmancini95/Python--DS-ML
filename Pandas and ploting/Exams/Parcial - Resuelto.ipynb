6)  Escribir  una  funci  ́oncontarque  reciba  una  lista  de  cadenas  y  devuelva  un  diccionario  que  contengauna  entrada  por  cada  cadena  en  la  lista  y  bajo  cada  entrada  guarde  la  cantidad  de  apariciones  deesa  cadena  en  la  lista  original.  Por  ejemplo,contar(Titulos)deber  ́ıa  devolver{’Mlle.’:  2,  ’Rev.’:  8,’Ms.’:  2,  ’Col.’:  4,  ’Countess.’:  1,  ’Sir.’:  1,  ’Mrs.’:  197,  ’Miss.’:  260,  ’Capt.’:  1,  ’Dona.’:  1,  ’Jonkheer.’:  1,’Master.’:  61,  ’Mr.’:  757,  ’Dr.’:  8,  ’Mme.’:  1,  ’Lady.’:  1,  ’Don.’:  1,  ’Major.’:  2}

```python
def contar(x): #se le ingresa una lista de cadenas
    d={}
    for i in x:
        if(i in d):
            d[i]=d[i]+1
        else:
             d[i]=1
    return d  #devuelve dicc que contiene una entrada por c/ cadena en la lista y abajo de cada entrada guarde la cantidad de apariciones deesa cadena en la lista original
```
