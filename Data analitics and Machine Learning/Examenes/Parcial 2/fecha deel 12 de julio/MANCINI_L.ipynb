#5. Búsqueda de la profundidad óptima
### En el ejercicio anterior usted calculó para una profundidad máxima de 3 nodos, la exactitud un árbol de decisión sobre los conjuntos train y test.
### Repita el procedimiento para valores de _max_depth_ entre 2 y 20, guardando las exactitudes correspondientes.

````python
l_train=list()
l_test=list()
for i in range(2,21):
    clf_rf = RandomForestClassifier(criterion='entropy', max_depth=i, random_state=12 )
    clf_rf.fit(X_train, y_train)
    clf_rf.fit(X_train,y_train)
    l_train.append(clf_rf.score(X_train,y_train))
    l_test.append(clf_rf.score(X_test,y_test))
````

### Grafique la exactitud para ambos conjuntos en función de la profundidad máxima.
### ¿Cúal es el mejor valor de max_depth? ¿porque? Calcule la matriz de confusion para el caso de max_depth seleccionado e interpretela.

````python
df_rf_scores=pd.DataFrame()
df_rf_scores["Max_Depth"]=range(2,21)
df_rf_scores["Train_Score"]=l_train
df_rf_scores["Test_Score"]=l_test
````

````python
%matplotlib inline
sns.factorplot(x='Max_Depth', y='Train_Score', data=df_rf_scores,color="green")
sns.factorplot(x='Max_Depth', y='Test_Score', data=df_rf_scores,color='red')
````

````python
best_score=max(df_rf_scores[df_rf_scores.Train_Score<0.99].Test_Score)
best_depth=np.array(df_rf_scores[(df_rf_scores.Test_Score==best_score)].Max_Depth).min()
print("El mejor valor que se puede elegir para la profundidad es el de ",best_depth," y da una exactitud en el score de test igual a,:",best_score)
print("obs: Se uso como parametro que una exactitud de entrenamiento mayor a 0.99 es overfitting, por lo tanto se descartaron las profundidades que presentaron esos valores. Luego de los que quedaron se busco el que tenia la mejor exactitud en la prueba de testeo, pero como habia dos profundidades con mismo valor de exactitud de testeo, para desempatar se apelo al principio de La Navaja de Ockham, que dice: En igualdad de condiciones, la explicación más sencilla suele ser la más probable. Por esta razon se eligio el modelo que tiene menor profundidad, es decir el que tiene mas libertad. ")
````

# 6. Bonus Track. K-folding:
##### Lo que hicimos en los primeros puntos de este examen puede interpretarse como la selección de un parámerto (_k_ ó _max_depth_) en función de la exactitud de la clasificación sobre un conjunto de testeo. El parámetro elegido como óptimo depende de la partición en train y test elegida. Le hemos sugerido utilizar la partición que genera un _random-satate_ particular para evitar que el azar le de una partición con característica muy especiales que le impidan sacar conclusiones razonables en este exámen. En este item vamos a tratar de reducir la influencia del azar en la partición de los datos utilizando una técnica de k-folding, lo que permitirá elegir el valor del parámetro de una forma más estable.

#### El siguiente fragmento de código separa los datos en 25 _folds_ e itera sobre ellos, generando conjuntos 25 conjuntos diferentes X_train, X_test (con sus y_train e y_test asociados).

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=25)
X=cancer.data
y=cancer.target
for train_index, test_index in skf.split(X,y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    print(X_test.shape)
```

### Combine este fragmento con el código que escribió en los ejercicios anteriores para evaluar la **exactitud media** de cada valor del parámetro (_k_ y _max_depth_, respectivamente) sobre las diferentes particiones. Grafique los resultados obtenidos (exactitud media en función del parámetro).  ¿Qué conclusiones puede sacar de este ejercicio?

````python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=25)
X=cancer.data
y=cancer.target
score_total_train_rf=list()
score_total_test_rf=list()
for i in range(2,21):
    forest = RandomForestClassifier(criterion='entropy', max_depth=i, random_state=12 )
    score_aux_train_rf=list()
    score_aux_test_rf=list()
    for train_index, test_index in skf.split(X,y):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        forest.fit(X_train, y_train)
        score_aux_train_rf.append(forest.score(X_train, y_train))
        score_aux_test_rf.append(forest.score(X_test, y_test))
    score_total_train_rf.append(np.mean(score_aux_train_rf))
    score_total_test_rf.append(np.mean(score_aux_test_rf))
````

````python
df_forest=pd.DataFrame()
df_forest["Max_Depth"]=range(2,21)
df_forest["Train_Score"]=score_total_train_rf
df_forest["Test_Score"]=score_total_test_rf
````

````python
%matplotlib inline
sns.factorplot(x='Max_Depth', y='Train_Score', data=df_forest)
sns.factorplot(x='Max_Depth', y='Test_Score', data=df_forest,color='red')
````

````python
best_score=max(df_forest[df_forest.Train_Score<0.99].Test_Score)
best_depth=np.array(df_forest[(df_forest.Test_Score==best_score)].Max_Depth).min()
#best_depth=np.array(df_forest[(df_forest.Test_Score==best_score)].Max_Depth)
print("El mejor valor que se puede elegir para la profundidad es el de ",best_depth," y da un score de test igual a,:",best_score,". Se puede ver que sin hacer el kfolding, habiamos descartado algunas profundidades por suponer que presentaban overfitting, pero que ahora en promedio parecen no tenerlo. ")
print("Como conclusion se puede pensar que hacer el k folding sirvio, y es buen metodo para elegir parametros, pero tambien podemos pensar que talvez el costo algoritmico es muy alto y no justifica su uso ya que al final el valor que se obtuvo para la exactitud con el conjunto de testeo fue muy similar (podriamos apelar nuevamente al principio de la navaja de Ockham y decidir que preferimos usar el metodo mas sencillo")
````
