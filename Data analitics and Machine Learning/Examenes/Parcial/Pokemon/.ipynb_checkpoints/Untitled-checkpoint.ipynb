```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Pokemon.csv", index_col=0)
df
```

SCATTER PLOT

```python

sns.lmplot(x="Attack",y="Defense",data=df) #es Attack vs Defende
```

```python
sns.lmplot(x="Attack",y="Defense",data=df,hue="Stage", fit_reg=False)#agregando el hue es como que puedo representar una tercera columna
#si quisiera las lineas le saco el  fit_reg=False
```

POKEMON CON MAS ATTACK Y DEFENSE

```python
sum_attack_defense=df.Attack+df.Defense
df[sum_attack_defense==sum_attack_defense.max()]
```

SETTEO DE PALETA DE COLORES

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]

pkmn_abilities_colors = ['green',  #HP
                    'Red',  # ATTACK
                    'Blue',  # DEFENSE
                    'Yellow',  # SPEED ATTACK
                    'Brown',  # SPEED DEFENSE
                    '#A040A0',  # SPEED
                   ]
```

BOXPLOT

```python
# nuevo data frame sin el total, le stage y el booleano de legendary
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)

sns.boxplot(data=stats_df,palette=pkmn_abilities_colors)
plt.ylim(0,None)#si queremos ver mas de cerca cambiamos el none por el limite superior que queramos
```

VIOLIN PLOT

```python
# nuevo data frame sin el total, le stage y el booleano de legendary
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)

plt.figure(figsize=(10,6))
sns.violinplot(data=stats_df)
plt.ylim(0,None)#si queremos ver mas de cerca cambiamos el none por el limite superior que queramos
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='Type 1', y='Attack',order=[ 
"Grass",
"Fire",
"Water",
"Bug",
"Normal",
"Poison",
"Electric",
"Ground",
"Fairy",
"Fighting",
"Psychic",
"Rock",
"Ghost",
"Ice",
"Dragon"] , palette=pkmn_type_colors, data=df)
#si quiero asegurarme que los colores de la paleta coincidan con los violines del graficos, tambien tengo que ponerle el orden igual que en la paleta 
```

CALCULO QUIENES TIENEN MAS TOTAL; LEGENDARIOS Y NO LEGENDARIOS

```python
df[(df.Total) == (df.Total.max())] #el que tiene mas total
```

```python
df[(df.Total) == (df[df.Legendary == False].Total.max())] #lsoq ue tienen mas total de los no legendarios
```

DISTPLOT (tipo histograma)

```python
plt.figure(figsize=(12,8))
sns.distplot(df[df.Stage==1].Total,color="red",bins=10)
sns.distplot(df[df.Stage==2].Total,color="blue")
sns.distplot(df[df.Stage==3].Total,color="green",axlabel="TOTAL")
```

FACTORPLOT

```python
sns.factorplot(x="Type 1",y="Total",data=df,kind="swarm",hue="Stage",col="Legendary").set_xticklabels(rotation=90)
```

COUNTPLOT

```python
sns.countplot(x='Stage' , data= df)
```

```python
sns.countplot(x='Stage' , data= df,hue="Legendary")
```

```python
sns.factorplot(x='Stage' , data= df , hue='Type 1' , kind= "count")
```
