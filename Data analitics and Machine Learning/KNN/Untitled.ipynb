```python
import numpy.random as rdm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn import neighbors, datasets
```

1.  Usando numpy.random.multivariatenormal,  generar  2x250  muestras  aleato-rias  de  dos  normales  bivariadas  (cercanas  pero  diferentes)  y  graficarlas  concolores  diferentes.

```python
#1)
#a)
mean1 = [100, 80]
mean2 = [90, 70]
cov1 = [[4, 0], [0, 35]]
cov2 = [[31, 0], [0, 25]]
a = np.random.multivariate_normal(mean1, cov1, 476)
b = np.random.multivariate_normal(mean2, cov2, 520)
plt.scatter(a[:,0], a[:,1], c = 'r')
plt.scatter(b[:,0], b[:,1], c = 'g')
plt.legend(['a', 'b'])
```

2 .  Concatenar  ambas  muestras  y  generar  un  vector  de  etiquetas  adecuado(Obtenga  una  matriz  X  de  500x2  con  las  muestras,  y  un  vector  Y  de  500de  largo  con  las  etiquetas).

```python
X = np.concatenate((a,b))
#Creo vector Y condatos binarios 
Y = np.full(996, [1] * 476 + [2] * 520)
```

```python
#separo en train y test:

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X , Y, test_size=0.2)
```

3  Utilizar  el  clasificador  KNN  de  sklearn  con  este  conjunto,  como  en  elejemplo  visto  en  clase  (KNN.ipnb  de  esta  carpeta).

```python
n_neighbors=3
clf = neighbors.KNeighborsClassifier(n_neighbors,weights="distance") #
clf.fit(X_train, Y_train) #aca aplico el KNN
clf.score(X_test,Y_test)
```

```python
Y_pred=clf.predict(X_test)
```

```python
plt.subplot(1,2,1)
plt.scatter(X_test[:,0], X_test[:,1], c=Y_pred)
plt.subplot(1,2,2)
plt.scatter(X_test[:,0], X_test[:,1], c=Y_test)
```

```python
clf.score(X_test,Y_test)
```

```python
scores=list()
for i in [1,7,12,18]:
    n_neighbors=i
    clf = neighbors.KNeighborsClassifier(n_neighbors,weights="distance") #
    clf.fit(X_train, Y_train) #aca aplico el KNN
    scores.append( clf.score(X_test,Y_test))
```

```python
from matplotlib.colors import ListedColormap
%matplotlib inline
    
num_neighbors=1
step_size=0.2
x_min, x_max = X[:,0].min()-1, X[:,0].max()+1
y_min, y_max = X[:,1].min()-1, X[:,1].max()+1

xx, yy = np.meshgrid(np.arange(x_min, x_max, step_size), np.arange(y_min, y_max, step_size))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

#Creo el mapa de colores
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

#Ploteo tambien los puntos

plt.scatter(X[:,0], X[:,1], c=Y, cmap=cmap_bold)
plt.title('Clasificacin con KNN, con k={0}'.format(num_neighbors))
        
  
```
